```python
from fbprophet import Prophet
m1 = Prophet()

datos1 = datos1.rename(index=str, columns={"Ammount": "y", "Fecha_Final": "ds"})
# datos1["ds"]= datos1["ds"]/10000000
m1.fit(datos1)

m2 = Prophet()

datos2 = datos2.rename(index=str, columns={"Ammount": "y", "Fecha_Final": "ds"})
# datos2["ds"]= datos2["ds"]/10000000
m2.fit(datos2)
```

```python
future1 = m1.make_future_dataframe(periods=10)
future1
# display(future1)

future2 = m2.make_future_dataframe(periods=10)
# future
# display(future2.tail(10))
```

```python
forecast1 = m1.predict(future1)
forecast1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

forecast2 = m2.predict(future2)
forecast2[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```
